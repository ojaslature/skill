Here we can see that the last col i.e. species is of categorical value, we will perform different types of encoding techniques on it only


Ordinal Encoding: 
- Assiging integers based on their order from 1

```python
#using scikit-learn
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder()
ordinal=data.copy()
# Apply ordinal encoding to the 'Species' column
ordinal[['Species']] = encoder.fit_transform(data[['Species']])
ordinal.sample(10)
```

Label Encoding
- Assigns integer value from 0

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
label=data.copy()
label['Species']=encoder.fit_transform(data['Species'])
label.sample(10)
```

One Hot encoding
- Creates a new col and input a binary string value (true or false)

```python
#using numpy 
encoder=pd.get_dummies(data['Species'],prefix='flower')
onehot=pd.concat([data,encoder],axis=1)
onehot.sample(10)
```

Binary encoding
- converts inegers into bool , then creates binary col
- more memory efficient then one-hot

Note: its bool values so 
- 00=0 
- 01=1 
- 10=2

```python
from category_encoders import BinaryEncoder
encoder=BinaryEncoder(cols=['Species'])
binary=encoder.fit_transform(data)
binary.sample(10)
```

Target Encoding(Mean Encoding)
- first groups all the same unique values in the encoding fature i.e. 3
- finds the different mean of the grp and assigns that instaed of the name here

```python
from category_encoders import TargetEncoder
y=np.array(data['Petal.Width'])
encoder=TargetEncoder(cols=['Species'])
target=encoder.fit_transform(data,y)
target.sample(10)
```

Frequency Encoding
- replaces categories with their frequency in the dataset
- applying normalisation to return proportion instead of count

```python
freq=data.copy()
feq_map=freq['Species'].value_counts(normalize=True).to_dict()
freq['freqency']=freq['Species'].map(feq_map)
freq.sample(10)
```

Hash Encoding
- applies hash function to categories , creating a fixed number of features
- n_component = 1.5/2 times the unique values in the target feature
- weird if used only with 3 distinct features


```python
from category_encoders import HashingEncoder
encoder=HashingEncoder(cols=['Species'],n_components=6)
hash=encoder.fit_transform(data)
hash.sample(10)
```

Leave One out Encoding
- varient of target encoding that prevents overfitting
- different from mean encoder as it does not look at the current row its working on

```python
from category_encoders import LeaveOneOutEncoder
encoder=LeaveOneOutEncoder(cols=['Species'])
leaveoneout=encoder.fit_transform(data,data['Petal.Width'])
leaveoneout.sample(10)
```

Grouping based on Frequency

```python
def frequency_based_grouping(data, column, threshold=0.05):

    value_counts = data[column].value_counts(normalize=True)
    mapping = {}
    for category, freq in value_counts.items():
        if freq < threshold:
            mapping[category] = 'Other'
        else:
            mapping[category] = category
    
    # Apply mapping
    data[f"{column}_grouped"] = data[column].map(mapping)
    return data
data=frequency_based_grouping(data,'Species')
display(data.sample(10))
```
